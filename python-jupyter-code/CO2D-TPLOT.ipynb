# Taylor plot (notebook for plotting CO2-diamond FTIR-data)
This is **Jupyter** notebook with embeded **Python (MatPlotLib + NumPy)** script. Script generates plots from csv-format tables. It is used to plot $CO_2$-diamond FTIR-data.

Options and loading data:

```python
import matplotlib.colors as colors
import numpy as np
import pandas as pd #Useful for work with arrays
import matplotlib.pyplot as plt #For creating plots
import matplotlib.pylab as pylab #For pylab.rcParams.update(params)
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator) #For ticks customisation
params = {'axes.labelsize': 'x-large', #global settings for label size
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params) #load global settings for labels
import math #Import for math.exp
from pathlib import Path #used for creating an output directory if it doest not exist.

# For matplotlib-label-lines

from labellines import labelLine, labelLines


path ='/Users/meguka/Dropbox/CO2D-project-github/results/tables/' #Path to the data
destination ='/Users/meguka/Dropbox/CO2D-project-github/results/spectra-plots/' #Path to export

#Load datsets for FN7114 diamond
Acenter = pd.read_table(path+'Acenter-from-FN7114 line across C.csv',  sep = ",", index_col=0)
Bcenter = pd.read_table(path+'Bcenter-from-FN7114 line across C.csv',  sep = ",", index_col=0)
```

```python
#Create part of viridis colormap to be as points 43-56
def truncate_colormap(cmapn, minval=0, maxval=1.0, n=13):
    new_cmapn = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmapn.name, a=minval, b=maxval),
        cmapn(np.linspace(minval, maxval, n)))
    return new_cmapn

cmapn = plt.get_cmap('viridis')
new_cmapn = truncate_colormap(cmapn, 0.76755, 1)
```

Define a nitrogen aggregation isotherm function:

```python
#DEFINE FUNCTION
def Calculate_Bfraction (Bfrac0, Ntot, tauSec, tempC): # Calculates Bfraction from Ntot and Bfrac0 after tauSec, tempC exposure.
    #Bfrac0 = initial B fraction from 0 to 1; Ntot = total nitrogen in atomic ppm; tauSec = time of exposure in seconds; tempC = temperature of exposure in C;   
    EaR = 8.16*10**4 #Set Ea/R ration in K
    Aarr=math.exp(13.51) #Set pre-exponential factor
    tempK=tempC+273.15 #Calculate temperature in K
    k = Aarr*math.exp(-EaR/tempK)
    Bfrac = 1-1/(Ntot*tauSec*k+1/(1-Bfrac0)) #Calculate new Bfraction 
    return Bfrac #Return new Bfraction

#SET PARAMETERS FOR ISOTHERMS
time1 = 0.5
time2 = 1.5
SecGa = 3.154*10**16 #seconds in one GigaAnnum
temperatures = range(1050, 1300, 50)
Ntot_Isotherm = np.linspace(1, 10000, 10000) #Create values of Total N for Isotherms
colors = np.linspace(0.8, 0.4, 6) #Colors for curves
Bfrac0 = 0 # initial B fraction

fig, ax = plt.subplots(1, 1, figsize=[6.5,5], constrained_layout=True) #Create figure and first axis

cm = Acenter[" Number_in_fun_name"].iloc[43:] #set for colormap
NitrogenTotal=(Bcenter[" height"].iloc[43:]+Acenter[" height"].iloc[43:]) #Total atomic nitrogen
Bfraction=Bcenter[" height"].iloc[43:]/NitrogenTotal #B fraction in atomic nitrogen
Afraction=Acenter[" height"].iloc[43:]/NitrogenTotal #A fraction in atomic nitrogen     
# annealing_time = 600
# annealing_temp = 2100
# # ChangedBfrac = Calculate_Bfraction(Bfraction, NitrogenTotal,annealing_time,annealing_temp) #Transform Bfrac
# # ChangedPlot = ax.scatter(ChangedBfrac*100, NitrogenTotal, label='After %s s at %s C' % (annealing_time, annealing_temp), marker="+", color='k', edgecolors='k', s=40) #Create scatter 

#TIMES
i = 0
for temp in temperatures:
    time = time1
    TauSec = time*SecGa
    IsothermBfraction1 = Calculate_Bfraction(Bfrac0, Ntot_Isotherm, TauSec, temp)
    ax.plot(IsothermBfraction1*100, Ntot_Isotherm, color='k', label="%s Ga, %s °C" % (time, temp)) #Plot 
    
    time = time2
    TauSec = time*SecGa
    IsothermBfraction2 = Calculate_Bfraction(Bfrac0, Ntot_Isotherm, TauSec, temp)
    ax.plot(IsothermBfraction2*100, Ntot_Isotherm, color='k', linestyle='--', label="%s Ga, %s °C" % (time, temp)) #Plot
    
    ax.fill_betweenx(Ntot_Isotherm,IsothermBfraction1*100, IsothermBfraction2*100, alpha=0.2, color=str((colors[i])),step='post', zorder=3, rasterized=True) #Fill between values.
    i = i + 1

experimentalData = ax.scatter(Bfraction*100, NitrogenTotal, label='Experimental', c=cm, cmap=new_cmapn, edgecolors='k', s=50) #Experimental data
clb = fig.colorbar(experimentalData,ax=ax, ticks=[45, 50, 55]) #Create colorbar
clb.ax.set_title('Profile point') #add title above colorbar. size - font size
    
#CUSTOMIZE
# legend = ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
# legend.get_frame().set_alpha(0.5)
ax.set(ylabel='Total nitrogen concentration (atomic ppm)', xlabel='%B') 
ax.set_xlim(xmin=0, xmax=100)
ax.set_ylim(ymin=10, ymax=10000)
plt.yscale(value="log")

xvals = [7,18,28,40,50, 60, 70, 75, 80, 85]
labelLines(plt.gca().get_lines(),xvals=xvals,zorder=2 ,color='k') #Create labels on lines

#SAVE    
Path(destination).mkdir(parents=True, exist_ok=True) #create directory at destination if does not exist. Uses pathlib
plt.savefig(destination+'TaylorPlot'+'.png', bbox_inches='tight', dpi=1000)#Saves image.
plt.savefig(destination+'TaylorPlot'+'.jpg', bbox_inches='tight', dpi=1000)#Saves image.
plt.savefig(destination+'TaylorPlot'+'.pdf', bbox_inches='tight', dpi=1000)#Saves image.
```
