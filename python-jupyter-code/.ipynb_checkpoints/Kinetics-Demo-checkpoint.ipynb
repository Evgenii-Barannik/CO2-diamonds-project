```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style

t = np.linspace(0, 5, 50)
k = 0.02
A0 = 100
Aconc=A0/(k*t*A0+1)

fig, ax0 = plt.subplots()
ax0.scatter(t, Aconc, label="A")
ax0.scatter(t, (A0-Aconc)/2, label="B", color='red')
ax0.set(ylabel="Concentration, %", xlabel="Time")
ax0.set_xlim(xmin=0)
ax0.set_ylim(ymin=0)
ax0.legend()
plt.savefig('Second-order-kinetics-plot.png', bbox_inches='tight')
```
