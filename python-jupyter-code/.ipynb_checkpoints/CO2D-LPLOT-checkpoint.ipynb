Make a nitrogen aggregation plot:

```python
fig, axs = plt.subplots(1, 2, figsize=(11, 4), sharey=True, sharex=True, constrained_layout=True)
axs[0].scatter(V1[" Area"]+V2[" Area"]+V3[" Area"], V5[" Area"]+ V6[" Area"], c=cm1, cmap='viridis', edgecolors='k', s=60) #Create scatter with panda data on ax0. SUPPORTS MATH IN THE INPUT!
p = axs[1].scatter(V1[" Area"]+V2[" Area"], V5[" Area"]+ V6[" Area"], c=cm1, cmap='viridis', edgecolors='k', s=60) #Create scatter with panda data on ax1. SUPPORTS MATH IN THE INPUT!
clb = fig.colorbar(p,ax=axs) #axs here make to draw colorbar for whole figure
axs[0].set(xlabel='CO$_2$ 3b + 3a area (cm$^{-2}$)', ylabel='CO$_2$ 2b + 2a area (cm$^{-2}$)')
axs[1].set(xlabel='CO$_2$ 3b area (cm$^{-2}$)')
clb.ax.set_title('Profile point') #add title above colorbar. size - font size

#linear regression:
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(V1[" Area"]+V2[" Area"]+V3[" Area"],V5[" Area"]+ V6[" Area"])
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(V1[" Area"]+V2[" Area"], V5[" Area"]+ V6[" Area"])
r2_1 = (r_value1*r_value1) #Calculate r^2
r2_2 = (r_value2*r_value2)

x = np.linspace(350,700,2) #Create two points on 350 and 700. For line plotting
axs[0].plot(x, intercept1 + slope1*x, '--k', label='Linear fit: \nslope = %.4f \nintercept = %.4f \n$r^2=%.4f$'  %  (slope1, intercept1, r2_1)) # .3f set 3 digits after.
axs[0].legend(loc='upper left') #Setting location with a vector.

axs[1].plot(x, intercept2 + slope2*x, '--k', label='Linear fit: \nslope = %.4f \nintercept = %.4f \n$r^2=%.4f$' %  (slope2, intercept2, r2_2))
axs[1].legend(loc='upper left') #Setting location with a vector.

axs[0].text(0.9, 0.1, 'a', size=30, transform=axs[0].transAxes) #Add letter. transform=ax.transAxes uses internal axes coordinates.
axs[1].text(0.9, 0.1, 'b', size=30, transform=axs[1].transAxes) #Add letter. transform=ax.transAxes uses internal axes coordinates.

axs[0].yaxis.set_major_locator(MultipleLocator(20)) #Set minor ticks
axs[0].yaxis.set_minor_locator(MultipleLocator(10)) #Set minor ticks
axs[0].xaxis.set_minor_locator(MultipleLocator(50))

Path(destination).mkdir(parents=True, exist_ok=True) #create directory at destination if does not exist. Uses pathlib
plt.savefig(destination+'FN7114 line across CO2 30 microns linear fit comparison.pdf', dpi=1000, bbox_inches='tight')#Saves image. dpi (dot per inch) sets resolution. bbox_inches = 'tight' makes images tightly cropped
plt.savefig(destination+'FN7114 line across CO2 30 microns linear fit comparison.png',  dpi=1000,bbox_inches='tight')#Saves image. dpi (dot per inch) sets resolution. bbox_inches = 'tight' makes images tightly cropped
plt.savefig(destination+'FN7114 line across CO2 30 microns linear fit comparison.jpg',  dpi=1000, bbox_inches='tight')#Saves image. dpi (dot per inch) sets resolution. bbox_inches = 'tight' makes images tightly cropped
```
