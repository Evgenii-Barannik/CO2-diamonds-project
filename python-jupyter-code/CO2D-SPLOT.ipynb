# Spectra plot (notebook for plotting CO2-diamond FTIR-data)
This is **Jupyter** notebook with embeded **Python (MatPlotLib + NumPy)** script. Script generates spectra images from csv-format tables.
It is used to plot $CO_2$-diamond FTIR-data.

Set OPTIONS: (Check and edit here)

```python
path ='/Users/meguka/Dropbox/CO2D-project-github/' #Path to the project
destination ='/Users/meguka/Dropbox/CO2D-project-github/results/spectra-plots/' #Path to export images
```

Import packages:

```python
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt
import matplotlib.style
from pathlib import Path #used for creating an output directory if it doest not exist.
import matplotlib.pylab as pylab #For pylab.rcParams.update(params)
params = {'axes.labelsize': 'x-large', #global settings for label size
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params) #load global settings for label size
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator) #For ticks
import pandas as pd
```

Define function to save:

```python
def save():
    subname = name.split(".")[0] #Takes string from filename up to "." symbol
    Path(destination).mkdir(parents=True, exist_ok=True) #create directory at destination if does not exist. Uses pathlib
    plt.savefig(destination+subname+'.pdf', dpi=1000, bbox_inches='tight')#Saves image. dpi (dot per inch) sets resolution. bbox_inches = 'tight' makes images tightly cropped
    plt.savefig(destination+subname+'.png', dpi=1000, bbox_inches='tight')#Saves image. dpi (dot per inch) sets resolution. bbox_inches = 'tight' makes images tightly cropped
    plt.savefig(destination+subname+'.jpg', dpi=1000, bbox_inches='tight')#Saves image. dpi (dot per inch) sets resolution. bbox_inches = 'tight' makes images tightly cropped
```

Load spectrum files and plot:

```python
name='FN7114Profile#2-31-01-20200003-V2-Fitted.dat' #Dataset file
fullpath=path+'/results/edited-spectra/'+name #Full path to the file
x, y, F1, F2, F3, F4, Sum, Dif = np.loadtxt(fullpath, unpack = True, delimiter = ' ') #Load data into columns. Name columns x, y, F1, F2, F3, F4, Sum, Dif
x, CO2A = np.loadtxt(path+'data/background-spectra/Pure CO2 + H2O background A(w).dat', unpack = True, delimiter = ' ')
fig = plt.figure(figsize=[5,6], constrained_layout=True ) #Create figure container.
gs = fig.add_gridspec(3, 1, height_ratios=[4, 1, 1], hspace=0) #Create grid for the figure container. 
ax0 = fig.add_subplot(gs[0, 0]) #define ax0 as the axes on the first (top) gridspace
ax1 = fig.add_subplot(gs[1, 0]) #define ax1 as the axes on the second (mid) gridspace
ax2 = fig.add_subplot(gs[2, 0]) #define ax2 as the axes on the third (low) gridspace

ax0.set(xlabel='Wavenumber (cm$^{-1}$)', ylabel='Absorbance coefficient (cm$^{-1}$)')
ax0.set_xlim(xmin=620, xmax=700)
ax0.set_ylim(ymin=0, ymax=7.5)
ax0.plot(x, y, 'black', label='experimental') #Create plot on ax0.
ax0.plot(x, F2, 'red', label='CO$_2$ ν2a, phase I') #Create plot on ax0.
ax0.plot(x, F3, 'maroon', linestyle='-.', label='CO$_2$ ν2b, phase I') #Create plot on ax0.
ax0.plot(x, F4, 'gray', label='CO$_2$ gas') #Create plot on ax0.
ax0.plot(x, F1, '--k', label='linear baseline') #Create plot on ax0.
ax0.legend(prop={'size': 10})
ax0.xaxis.set_minor_locator(MultipleLocator(10)) #Set minor ticks

ax1.plot(x, Dif, 'black', label='residue') #Create plot on ax1.
ax1.set_xlim(xmin=620, xmax=700)
ax1.set_ylim(ymin=-1, ymax=1)
ax1.tick_params(axis='y', which='both')
ax1.tick_params(axis='x', which='both', labelbottom=False, labelleft=False)
ax1.legend()
ax1.xaxis.set_minor_locator(MultipleLocator(10)) #Set minor ticks
#ax2.set(ylabel='Absorbance')
ax2.plot(x, CO2A, 'gray', label='CO$_2$ gas experimental') #Create plot on ax2.
ax2.set_xlim(xmin=620, xmax=700)
ax2.set_ylim(ymin=0, ymax=0.9)
ax2.tick_params(axis='x', which='both', direction='out', labelbottom=False, labelleft=False)
ax2.tick_params(axis='y', which='both', length=0, labelbottom=False, labelleft=False)
ax1.hlines(0, 600, 2500, linewidth=0.5) #Plot horizontal line
ax2.legend()
ax2.xaxis.set_minor_locator(MultipleLocator(10)) #Set minor ticks
save()

#Export data with Pandas
dict = {'wavenumber': x, 'experimental': y, 'CO2 v2a phase I': F2, 'CO2 v2b phase I' : F3, 'CO2 gas' : F4, 'linear baseline' : F1, 'residue' : Dif, 'CO2 gas experimental' : CO2A}  
export = pd.DataFrame(dict) 
export.to_csv(r'FN7114-V2.csv')
```

```python
name='FN7114Profile#2-31-01-20200003-V3-Fitted.dat'
fullpath=path+'/results/edited-spectra/'+name
x, y, F1, F2, F3, F4, F5, F6, F7, Sum, Dif = np.loadtxt(fullpath, unpack = True, delimiter = ' ') #load data from file
x, CO2A = np.loadtxt(path+'data/background-spectra/Pure CO2 + H2O background A(w).dat', unpack = True, delimiter = ' ') #load data from file
fig = plt.figure(figsize=[5,6], constrained_layout=True ) #Create figure container.
gs = fig.add_gridspec(3, 1, height_ratios=[4, 1, 1], hspace=0) #Create grid for the figure container. 
ax0 = fig.add_subplot(gs[0, 0]) #define ax0 as the axes on the first (top) gridspace
ax1 = fig.add_subplot(gs[1, 0]) #define ax1 as the axes on the second (mid) gridspace
ax2 = fig.add_subplot(gs[2, 0]) #define ax2 as the axes on the third (low) gridspace

ax0.set(xlabel='Wavenumber (cm$^{-1}$)', ylabel='Absorbance coefficient (cm$^{-1}$)')
ax0.set()            
ax0.set_xlim(xmin=2250, xmax=2500)
ax0.set_ylim(ymin=0, ymax=20)
ax0.plot(x, y, 'black', label='experimental') #Create plot on ax0.
ax0.plot(x, F2+F3, 'red', label='CO$_2$ ν3b, phase I') #Create plot on ax0.
ax0.plot(x, F5,  linestyle='-.', color="darkred", label='$^{13}$CO$_2$ ν3b, phase I') #Create plot on ax0.
ax0.plot(x, F4, linestyle=(0, (1, 1)), label='ν3a') #Create plot on ax0. (0, (1, 1)) is a densly dotted linestyle.
ax0.plot(x, F6, 'gray', label='CO$_2$ gas') #Create plot on ax0.
ax0.plot(x, F1, '--k', label='linear baseline') #Create plot on ax0.
ax0.legend(prop={'size': 10})
ax0.arrow(2300, 1.5, 0, 5, head_width=5, head_length=1, color='black') #Arrow plotting
ax0.xaxis.set_minor_locator(MultipleLocator(25)) #Set minor ticks

ax1.plot(x, Dif, 'black', label='residue') #Create plot on ax1.
ax1.hlines(0, 600, 2500, linewidth=0.5) #Plot horizontal line
ax1.set_xlim(xmin=2250, xmax=2500)
ax1.set_ylim(ymin=-0.5, ymax=0.5)
ax1.tick_params(axis='y', which='both')
ax1.tick_params(axis='x', which='both', labelbottom=False, labelleft=False)
ax1.legend()
ax1.xaxis.set_minor_locator(MultipleLocator(25)) #Set minor ticks

# ax2.set(ylabel='Absorbance')
ax2.plot(x, CO2A, 'gray', label='CO$_2$ gas experimental') #Create plot on ax2.
ax2.set_xlim(xmin=2250, xmax=2500)
ax2.set_ylim(ymin=0, ymax=1.5)
ax2.tick_params(axis='x', which='both', direction='out', labelbottom=False, labelleft=False)
ax2.tick_params(axis='y', which='both', length=0, labelbottom=False, labelleft=False)
ax2.legend()
ax2.xaxis.set_minor_locator(MultipleLocator(25)) #Set minor ticks

left, bottom, width, height = [0.23, 0.7, 0.2, 0.2] #Inset plotting
ax3 = fig.add_axes([left, bottom, width, height])
ax3.set_xlim(xmin=2280, xmax=2320)
ax3.yaxis.set_major_locator(MultipleLocator(0.5))

ax3.set_ylim(ymin=0, ymax=1)
ax3.plot(x, y, 'black') 
ax3.plot(x, F5, linestyle='-.', color="darkred") 
ax3.plot(x, F2+F3, 'red') 
ax3.plot(x, F6, 'gray') 
ax3.plot(x, F1, '--k')
ax3.tick_params(axis='both', which='major', labelsize=10)
save()

#Export data with Pandas
dict = {'wavenumber': x, 'experimental': y, 'CO2 v3b phase I' :F2+F3, '13CO2 v3b phase I' :F5, 'v3a': F4, 'CO2 gas' : F6, 'linear baseline' : F1, 'residue' : Dif, 'CO2 gas experimental' :CO2A}  
export = pd.DataFrame(dict) 
export.to_csv(r'FN7114-V3.csv')
```
